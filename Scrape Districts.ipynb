```python
from requests import get
from requests.exceptions import RequestException
from contextlib import closing
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://www.boston.gov/departments/police#districts'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
soup
```

```python
# Find the number of districts
results = soup.find_all('div', class_='cd-c')
```

```python
district_codes = []
for result in results:
    # Retrieve the thread title
    district = result.find('div', class_='cd-t')
    # Access the thread's text content
    print(district.text)
    district_codes.append(district.text.strip())

    
```

```python
district_codes
```

```python
district_names = []
for result in results:
    # Retrieve the thread title
    district_name = result.find('div', class_='cd-d')
    # Access the thread's text content
    print(district_name.text)
    district_names.append(district_name.text.strip())
```

```python
district_names
```

```python
act_codes = []

for i in range(len(district_codes)):
    x = district_codes[i]
    if x.find("District") > -1:
        act_codes.append(i)
        
act_codes        
```

```python
district_codes2 = []
district_names2 = []

for i in range(len(act_codes)):
    x = act_codes[i]
    district_codes2.append( district_codes[x])
    district_names2.append(district_names[x])
```

```python
district_codes2
```

```python
#district_codes2[0].split('and')[1]
district_names2
```

```python
removal_dn = []
removal_dc = []
district_codes4 = []
district_names4 = []

for i in range(len(district_codes2)):
    #print(i)
    x = district_codes2[i]
    if len(x.split('and')) > 1: 
        for j in range(len(district_codes2[i].split('and'))): 
            district_codes4.append(district_codes2[i].split('and')[j])
            district_names4.append(district_names2[i])
            
        removal_dc.append(district_codes2[i])
        removal_dn.append(district_names2[i])
```

```python
removal_dc
removal_dn
```

```python
district_codes4
district_names4
```

```python
for i in range(len(removal_dc)):
    district_codes2.remove(removal_dc[i])
    district_names2.remove(removal_dn[i])
```

```python
for i in range(len(district_codes4)):
    district_codes2.append(district_codes4[i])
    district_names2.append(district_names4[i])
            
```

```python
for i in range(len(district_codes2)):
    x = district_codes2[i]
    if x.find("District") == -1:
        district_codes2[i] = "District " + district_codes2[i]
```

```python
d = {'District Codes':district_codes2,'District Names':district_names2}
d

df = pd.DataFrame(d)
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
```

```python

zip_df = pd.io.parsers.read_csv('Data/Boston_ZipCodes.csv', dtype='str')
zip_df
```

```python
zip_df.columns
zip_df.columns = zip_df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
zip_df.columns
```

```python
dist_list = zip_df['district'].tolist()
dist_list
```

```python
zip_list = zip_df['zip_code'].tolist()
zip_list
```

```python
#White
white_pop = []
for i in range(len(zip_list)):
    if zip_list[i] in ['02112', '02117', '02123', '02133', '02137']:
        white_pop.append('NA')
    else: 
        url="https://api.census.gov/data/2017/acs/acs5?key=9a52ef9938baee032bc5769ef064a1a1fd9f7332&get=B01001A_001E&for=zip%20code%20tabulation%20area:" + str(zip_list[i])
        JSONContent = requests.get(url).json()
        white_pop.append(JSONContent[1][0])
        print(url)

white_pop
```

```python
#Black/African American

aa_pop = []

for i in range(len(zip_list)):
        if zip_list[i] in ['02112', '02117', '02123', '02133', '02137']:
            aa_pop.append('NA')
        else:
            url="https://api.census.gov/data/2017/acs/acs5?key=9a52ef9938baee032bc5769ef064a1a1fd9f7332&get=B01001B_001E&for=zip%20code%20tabulation%20area:" + str(zip_list[i])
            JSONContent = requests.get(url).json()
            aa_pop.append(JSONContent[1][0])
            print(url)
            

aa_pop
```

```python
d = {'Zip Code':zip_list,'District':dist_list, "White Pop":white_pop, "Afr Am Pop":aa_pop, "Asian Pop":asian_pop, "Native Am Pacific Pop":napi_pop, "Other Pop":other_pop,"Multi Race Pop":multi_pop }
d

race_dem_df = pd.DataFrame(d)
race_dem_df 
```
